```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
sims = input('How many simulations would you like to run?  ') 
i = 0
wins = 0
losses = 0

while i < int(sims):
    die_1 = random.randint(1,6)
    die_2 = random.randint(1,6)
    tot_1 = die_1 + die_2
    if tot_1 == 7:
        wins = wins + 1
    elif tot_1 == 11:
        wins = wins + 1
    elif tot_1 == 12:
        losses = losses + 1
    elif tot_1 == 2:
        losses = losses + 1
    else:
        x = 0
        while x != 1:
            die_1 = random.randint(1,6)
            die_2 = random.randint(1,6)
            tot_2 = die_1 + die_2
            if tot_2 == tot_1:
                wins = wins + 1
                x = 1
            elif tot_2 == 7:
                losses = losses + 1
                x = 1
    i = i + 1
    
    
totals = [wins, losses]
x_axis = np.arange(len(totals))
colors = ['green', 'red']
explode = (0, 0.1)
labels = ['Wins', 'Losses']
plt.bar(x_axis, totals, color=colors, align="center")
plt.title("# of Wins/Losses")
plt.grid()
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, labels)
for i, v in enumerate(totals):
    plt.text(i-0.1, max(totals)*0.5, str(totals[i]), color='white', fontsize=18)
plt.figure(figsize=(10,7))
plt.pie(totals, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.title("% of Wins/Losses")        

```
